```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

dataA = np.loadtxt("Sol.log", delimiter=",")
dataB = np.loadtxt("Sol_ref.log", delimiter=",")
```

```python
N = len(dataA)
x = np.linspace(1/N,(N-1)/N,N)
y = np.linspace(0,1,N)
X, Y = np.meshgrid(x, y)
Z = dataA

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u(x,y)')
ax.set_title("Jacobi2D_C,  N=8")

plt.show()
fig.savefig("plots/Jacobi2D_C.png")
```

```python
Z = dataB

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u(x,y)')
ax.set_title("Jacobi2D_A,  N=8")

plt.show()
fig.savefig("plots/Jacobi2D_A.png")
```
